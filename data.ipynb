```python
# import required modules

import pandas as pd
from pathlib import Path
import numpy as np
```

```python
# read and review original dataset

df = pd.read_csv(Path("Resources/american_bankruptcy_dataset.csv"))
df
```

```python
# count number of labeled status values

df["status_label"].value_counts()
```

```python
# transform string value to numerical value

df["status"] = np.where(df["status_label"] == 'alive', 0, 1)
```

```python
# verify counts matches with original column

df["status"].value_counts()
```

```python
# set column names

columns = ["company_name", "status_label", "year", "current_assets", "cost_of_goods_sold", "depreciation_and_amortization", "EBITDA", "inventory", "net_income", "total_receivables", "market_value", "net_sales", "total_assets", "total_long_term_debt", "EBIT", "gross_profit", "total_current_liabilities", "retained_earnings", "total_revenue", "total_liabilities", "total_operationg_expenses", "status"]
df.columns = columns
```

```python
# drop categorical values and review data

df = df.drop(columns=["company_name", "status_label", "year"])
df
```

```python
# save cleaned data to new csv file

df.to_csv(Path("Resources/relabeled_data.csv"), index=False)
```

```python
# import and review new csv file

renamed_df = pd.read_csv(Path("Resources/relabeled_data.csv"))
renamed_df
```

```python
# import modules required for data modification

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from imblearn.combine import SMOTEENN
```

```python
# split data into X and y
X = renamed_df.drop(columns="status")
y = renamed_df["status"]
display(X.tail())
display(y.tail())
```

```python
# split data to train and test

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# resample training data

smoteen_resampler = SMOTEENN(random_state=1)
X_train_resampled, y_train_resampled = smoteen_resampler.fit_resample(X_train, y_train)
display(y_train_resampled.value_counts())
```

```python
# apply standard scaler

X_scaler = StandardScaler()
X_train_resampled_scaled = X_scaler.fit_transform(X_train_resampled)
X_test_scaled = X_scaler.fit_transform(X_test)
display(X_train_resampled_scaled.shape)
display(y_train_resampled.shape)
display(X_test_scaled.shape)
display(y_test.shape)
```

```python
# combine and save training file

train = pd.concat([pd.DataFrame(X_train_resampled_scaled), pd.Series(y_train_resampled)], axis=1)
train.columns = renamed_df.columns
train.to_csv(Path("Resources/training_dataset.csv"), index=False)
train
```

```python
# combine and save training file

test = pd.concat([pd.DataFrame(X_test_scaled), pd.Series(y_test).reset_index(drop=True)], axis=1)
test.columns = renamed_df.columns
test.to_csv(Path("Resources/testing_dataset.csv"), index=False)
test
```

```python
# create !!NOT RESAMPLED!! csv file

X_train_scaled = X_scaler.fit_transform(X_train)
train_not_resampled = pd.concat([pd.DataFrame(X_train_scaled), pd.Series(y_train).reset_index(drop=True)], axis=1)
train_not_resampled.columns = renamed_df.columns
train_not_resampled.to_csv(Path("Resources/training_dataset_not_resampled.csv"), index=False)
train_not_resampled
```

```python
# create undersampled csv file

from imblearn.under_sampling import RandomUnderSampler

undersampler = RandomUnderSampler(random_state=1)
X_train_undersampled, y_train_undersampled = undersampler.fit_resample(X_train, y_train)


X_train_scaled_undersampled = X_scaler.fit_transform(X_train_undersampled)
train_undersampled = pd.concat([pd.DataFrame(X_train_scaled_undersampled), pd.Series(y_train_undersampled)], axis=1)
train_undersampled.columns = renamed_df.columns
train_undersampled.to_csv(Path("Resources/training_dataset_undersampled.csv"), index=False)
train_undersampled
```
