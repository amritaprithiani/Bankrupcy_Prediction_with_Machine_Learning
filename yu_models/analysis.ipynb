```python
# core imports

import pandas as pd
import numpy as np
from pathlib import Path

from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# import training/testing dataset

training_df = pd.read_csv(Path("../Resources/training_dataset_original.csv"))
testing_df = pd.read_csv(Path("../Resources/testing_dataset.csv"))

# split training dataset to X and y

X_train = training_df.drop(columns="status")
y_train = training_df["status"]
display(X_train.tail())

# split testing dataset to X and y

X_test = testing_df.drop(columns="status")
y_test = testing_df["status"]
display(X_test.tail())
```

### Class 1 precision

```python
from sklearn.svm import SVC

precision_model = SVC(kernel='rbf')
precision_model.fit(X_train, y_train)
precision_pred = precision_model.predict(X_test)

print(classification_report(y_test, precision_pred))
matrix_plot = ConfusionMatrixDisplay(confusion_matrix(y_test, precision_pred)).plot()
plt.show()
```

### Class 1 recall

```python
from imblearn.under_sampling import RandomUnderSampler

sampler = RandomUnderSampler()
X_resample, y_resample = sampler.fit_resample(X_train, y_train)

recall_model = SVC(kernel='rbf')
recall_model.fit(X_resample, y_resample)
recall_pred = recall_model.predict(X_test)

print(classification_report(y_test, recall_pred))
matrix_plot = ConfusionMatrixDisplay(confusion_matrix(y_test, recall_pred)).plot()
plt.show()
```

### Combine precision_pred and recall_pred into a dataframe

```python
pred_df = pd.concat([
    y_test,
    pd.Series(precision_pred),
    pd.Series(recall_pred)
], axis=1)
pred_df.columns = ['true', 'best_precision', 'best_recall']
pred_df
```

```python
pred_df['match'] = pred_df['best_recall'] - pred_df['best_precision']
```

```python
pred_df
```

```python
# when racall - precision == 1, overly ambiguous value, so make it 0
# else, use recall value

pred_df['combined'] = np.where(pred_df['match'] == 1, 0, pred_df['best_recall'])
```

```python
print(classification_report(pred_df['true'], pred_df['combined']))
matrix_plot = ConfusionMatrixDisplay(confusion_matrix(pred_df['true'], pred_df['combined'])).plot()
plt.show()
```
